```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

# Define constraints
x1 = np.linspace(0, 10, 100)  # Values for x1

# Constraint 1: 2x1 + 3x2 <= 12 -> x2 = (12 - 2x1) / 3
x2_c1 = (12 - 2 * x1) / 3

# Constraint 2: x1 + 2x2 <= 8 -> x2 = (8 - x1) / 2
x2_c2 = (8 - x1) / 2

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x1, x2_c1, label=r'$2x_1 + 3x_2 \leq 12$', color='blue')
plt.plot(x1, x2_c2, label=r'$x_1 + 2x_2 \leq 8$', color='green')

# Shading feasible region
plt.fill_between(x1, np.minimum(x2_c1, x2_c2), 0, where=(np.minimum(x2_c1, x2_c2) >= 0), color='gray', alpha=0.5)

# Axis settings
plt.xlim(0, 10)
plt.ylim(0, 5)
plt.xlabel('$x_1$ (Product A)', fontsize=12)
plt.ylabel('$x_2$ (Product B)', fontsize=12)
plt.title('Feasible Region for Problem 1', fontsize=14)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# Optimization using linprog
c = [-3, -4]  # Coefficients of the objective function (negative for maximization)
A = [[2, 3], [1, 2]]  # Coefficients of the inequalities
b = [12, 8]  # Right-hand sides of the inequalities
result = linprog(c, A_ub=A, b_ub=b, bounds=(0, None), method='highs')

# Optimal solution
print("Optimal solution:")
print(f"x1 (Product A) = {result.x[0]:.2f}")
print(f"x2 (Product B) = {result.x[1]:.2f}")
print(f"Maximum Profit = {-result.fun:.2f}")
```
